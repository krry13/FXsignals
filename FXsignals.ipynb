```python
import pandas as pd
import plotly.express as px
import numpy as np
import yfinance as yf

pd.set_option('display.max_rows',None)
```

```python
def get_data(tickers = 'AUDUSD=X', period = '1mo', interval = '15m'):
  return yf.download(tickers = tickers, period = period, interval = interval)

'''
^^^
Downloading financial data with yfinance

tickers: <typical fx pair name>=X i.e. AUDUSD=X

interval: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h,
1d, 5d, 1wk, 1mo, 3mo

period: 1d, 5d, 1mo, 3mo, 6mo,
1y, 2y, 5y, 10y, ytd, max
'''






def columns_manipulation(df_name, drop_col=['Volume','Adj Close'], separate = True):
  df = df_name.drop(drop_col, axis = 1)
  #df['Date'] = pd.to_datetime(df.index).date
  #df['Time'] = pd.to_datetime(df.index).time
  #df.set_index(['Date','Time'], inplace=True)
  df['ema20'] = df['Close'].ewm(span=20,min_periods=20,adjust=False,ignore_na=False).mean()
  df['ema50'] = df['Close'].ewm(span=50,min_periods=50,adjust=False,ignore_na=False).mean()




  return df

'''
^^^
Dropping "Volume" and "Adj Close" columns by default
Separating Date and Time from Datetime and making multindex from them

'''


fig = px.line(columns_manipulation(get_data()), y=['ema20','ema50'])
fig.show()










```
